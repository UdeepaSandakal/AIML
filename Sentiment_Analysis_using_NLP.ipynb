```python
import pandas as pd
```

```python
df = pd.read_csv('Tweets.csv')
```

```python
df = df[["airline_sentiment", "text"]]
```

```python
df.groupby('airline_sentiment').describe()
```

```python
import nltk
import string
import re
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords

nltk.download('stopwords')
nltk.download('punkt')
ps = PorterStemmer()

def clean_text(text):
    text = text.lower()
    text = re.sub(r'http.?://[^\s]+[\s]?', '', text)
    text = nltk.word_tokenize(text)
    y = []
    for i in text:
        if i not in stopwords.words('english'):
            y.append(i)
    text = y[:]
    y.clear()
    for i in text:
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df.head(3)
```

```python
# Applying "clean_text" function to "text" column and creating a new column "text_cleaned"
df['text_cleaned'] = df['text'].apply(clean_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizor = TfidfVectorizer(max_features=3000)
```

```python
# Generate TF-IDF vector representation of "text_cleaned" column and convert it to an array X
X = vectorizor.fit_transform(df['text_cleaned']).toarray()
```

```python
# Convert "airline_sentiment" column to an array Y
Y = df['airline_sentiment'].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape)
```

```python
# Train a multinomial Naïve Bayes classifier
nb_classifier = MultinomialNB()
nb_classifier.fit(X_train, y_train)
```

```python
# Predict sentiment of tweets in the test dataset using Naïve Bayes classifier
y_pred_nb = nb_classifier.predict(X_test)
```

```python
# Find the accuracy of Naïve Bayes classifier
accuracy_nb = accuracy_score(y_test, y_pred_nb)
print("Accuracy of Naïve Bayes classifier:", accuracy_nb)
```

```python
# Train a Random Forest classifier
model = RandomForestClassifier()

model.fit(X_train,y_train)

y_pred = model.predict(X_test)

print(accuracy_score(y_test,y_pred))
```
