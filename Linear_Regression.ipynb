#**1. Simple Linear Regression**


**Task**: Predict student score given the number of hours spent by the student for studies

**Input**: Hours studies

**Output**: Score

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("w1_student_scores.csv")
```

```python
df.head()
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(6, 4))
plt.plot(df['Hours'], df['Scores'], 'ro')
plt.title('Student Score vs Hours Studied')
plt.xlabel('Hours')
plt.ylabel('Scores')

plt.show()
```

**Find Coeffients**

calculate the mean of input and output


```python
xmean = np.mean(df['Hours'])
ymean = np.mean(df['Scores'])
```

calculate the terms needed for the numator and denominator of beta

```python
df['xycov'] = (df['Hours'] - xmean ) * (df['Scores'] - ymean )
df['xvar'] = (df['Hours'] - xmean ) ** 2
```

Calculate the coeffients

```python
slope = df['xycov'].sum() / df['xvar'].sum()
intercept = ymean - (slope * xmean )
print(f'slope = {slope}')
print(f'intercept = {intercept}')
```

Making prediction

```python
Scores_predicted = intercept + slope * df['Hours']
```

Compute the scores of a studnt who studied for 7 hours

```python
plt.figure(figsize=(6, 4))
plt.plot(df['Hours'], Scores_predicted)           # regression line
plt.plot(df['Hours'], df['Scores'], 'ro')         # scatter plot showing actual data
plt.title('Actual vs Predicted')
plt.xlabel('Hours')
plt.ylabel('Scores')

plt.show()
```

#**2. Multiple Linear Regression**

Task : Predict petrol consumption of a country given various factors such as tax, distance of highways and population with driving licence

Input

```python
df = pd.read_csv("w1_petrol_consumption.csv")
```

```python
df.head(5)
```

```python
df.shape
```

```python
df.describe().round(2)
```

```python
df.describe().round(2).T
```

```python
plt.figure(figsize=(6, 4))
plt.plot(df['Petrol_tax'], df['Petrol_Consumption'], 'ro')
plt.title('Petrol Tax vs Consumption')
plt.xlabel('Petrol_tax')
plt.ylabel('Petrol_Consumption')
plt.show()
```

Identify Outliers

```python
plt.figure(figsize=(6, 4))
plt.plot(df['Average_income'], df['Petrol_Consumption'], 'ro')
plt.title('Averag Income vs Petrol Consumption')
plt.xlabel('Average_income')
plt.ylabel('Petrol_Consumption')

plt.show()
```

Train the model

```python
Y = df['Petrol_Consumption']
X = df[['Average_income', 'Paved_Highways', 'Population_Driver_licence(%)', 'Petrol_tax']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
regressor.intercept_
```

```python
regressor.coef_
```

Make Prediction

```python
y_pred = regressor.predict(X_test)
```

```python
results = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(results)
```

Evaluating the Model

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
MAE = mean_absolute_error(y_test, y_pred)
MSE = mean_squared_error(y_test, y_pred)
RMSE = np.sqrt(MSE)

print(f'Mean Abloute error: {MAE: .2f}')
print(f'Mean Squared error: {MSE: .2f}')
print(f'Root Mean Squared error: {RMSE: .2f}')
```

$R^2$ score

```python
regressor.score(X_test, y_test)
```

```python
regressor.score(X_train, y_train)
```
